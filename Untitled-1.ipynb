```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os

```

```python
df = pd.read_excel("dirty_student_dataset (1).xlsx")
```

```python
df.isnull().sum()
```

```python
# 🎯 Learn
# Convert wrong types → numeric
# Invalid → becomes NaN (important)

df['screen_time'] = pd.to_numeric(df['screen_time'], errors='coerce')
```

```python
#  Clean Categorical Data

df['relationship'] = df['relationship'].str.strip().str.lower()
df['exercise'] = df['exercise'].str.strip().str.lower()

# scalable way 
# cat_cols = ['relationship', 'exercise']

# for col in cat_cols:
    # df[col] = df[col].str.strip().str.lower()
```

```python
# 6. Handle Missing Values (CORE)

num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
    
    # Numeric → MEDIAN (best practice)
```

```python
cat_cols = df.select_dtypes(include=['object', 'string']).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
    
    # ✅ Categorical → MODE
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python

# Ensure column is numeric
df['sleep_hours'] = pd.to_numeric(df['sleep_hours'], errors='coerce')

# Plot
sns.histplot(data=df, x='sleep_hours', kde=True)
plt.title("Sleep Hours Distribution")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot (Outliers)
sns.boxplot(x=df['stress_level'])
plt.title("Stress Level Boxplot")
plt.show()
```

```python
# 🧠 What you just did

# 👉 Used boxplot to detect:

# spread of data
# median
# outliers
# 📊 Your Graph Breakdown (clear)
# 🔹 Min ≈ 4
# 🔹 Q1 ≈ 5
# 🔹 Median ≈ 6
# 🔹 Q3 ≈ 7
# 🔹 Max ≈ 8

# 👉 Box = middle 50% data (5 → 7)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Style (big impact)
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x="sleep_hours",
    y="focus_level",
    s=80,                # point size
    color="#4C72B0",     # clean blue
    edgecolor="black"
)

plt.title("Sleep vs Focus", fontsize=14, weight="bold")
plt.xlabel("Sleep Hours", fontsize=12)
plt.ylabel("Focus Level", fontsize=12)

plt.tight_layout()
plt.show()




# sns.scatterplot(x='sleep_hours', y='focus_level', data=df)
# plt.title("Sleep vs Focus")
# plt.show()
```
